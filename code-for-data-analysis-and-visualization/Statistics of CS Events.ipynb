```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read time-event excel data
csevt=pd.read_excel('E:\. Disertasi S3 Bismillah\Olah Data\CS Lim Index V Based Event Data RM3.xlsx')
csevt=csevt.set_index('time_peak')
```

```python
#Statistic of CENS Event #3 days running mean\
print('Number of CS Event (D(0)JF(1) 1981 - D(0)JF(1) 2020)  : ', len(csevt), ' events')
print('Number of CS-CENS Event                               : ', len(csevt[csevt['censevt']==1]), ' events')
print('Number of CS-noCENS Event                             : ', len(csevt[csevt['censevt']==0]), ' events')
print('CENS Frequency per Season                             : ', len(csevt)/(2020-1980), ' events')
print('CENS Frequency per Season CS-CENS Event               : ', len(csevt[csevt['censevt']==1])/(2020-1980), ' events')
print('CENS Frequency per Season CS-noCENS Event             : ', len(csevt[csevt['censevt']==0])/(2020-1980), ' events')
print('CENS Index Average                                    : ', csevt['wsmagcs'].mean(), ' m/s')
print('CENS Index Average CS-CENS Event                      : ', csevt[csevt['censevt']==1]['wsmagcs'].mean(), ' m/s')
print('CENS Index Average CS-noCENS Event                    : ', csevt[csevt['censevt']==0]['wsmagcs'].mean(), ' m/s')
print('U CENS Index Average                                  : ', csevt['umagcs'].mean(), ' m/s')
print('U CENS Index Average CS-CENS Event                    : ', csevt[csevt['censevt']==1]['umagcs'].mean(), ' m/s')
print('U CENS Index Average CS-noCENS Event                  : ', csevt[csevt['censevt']==0]['umagcs'].mean(), ' m/s')
print('V CENS Index Average                                  : ', csevt['vmagcs'].mean(), ' m/s')
print('V CENS Index Average CS-CENS Event                    : ', csevt[csevt['censevt']==1]['vmagcs'].mean(), ' m/s')
print('V CENS Index Average CS-noCENS Event                  : ', csevt[csevt['censevt']==0]['vmagcs'].mean(), ' m/s')
print('Duration Total                                        : ', csevt['duration'].sum(), ' days')
print('Duration Total CS-CENS Event                          : ', csevt[csevt['censevt']==1]['duration'].sum(), ' days')
print('Duration Total CS-noCENS Event                        : ', csevt[csevt['censevt']==0]['duration'].sum(), ' days')
print('Duration Average                                      : ', csevt['duration'].mean(), ' days')
print('Duration Average CS-CENS Event                        : ', csevt[csevt['censevt']==1]['duration'].mean(), ' days')
print('Duration Average CS-noCENS Event                      : ', csevt[csevt['censevt']==0]['duration'].mean(), ' days')
```

```python
fig = plt.figure(figsize=(8,11))
plt.subplot(3, 1, 1)
countdurcs, bins_countdurcs = np.histogram(csevt[csevt['censevt']==1]['duration'], bins=24, range=(0.5,24.5))
countdurncs, bins_countdurncs = np.histogram(csevt[csevt['censevt']==0]['duration'], bins=24, range=(0.5,24.5))
plt.bar((bins_countdurcs[:-1] + bins_countdurcs[1:])/2, countdurcs/len(csevt[csevt['censevt']==1]), color='b', align='center', width=1, alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-CENS')
plt.bar((bins_countdurncs[:-1] + bins_countdurncs[1:])/2, countdurncs/len(csevt[csevt['censevt']==0]), color='orange', align='center', width=1, alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-noCENS')
plt.bar((bins_countdurcs[:-1] + bins_countdurcs[1:])/2, countdurcs/len(csevt[csevt['censevt']==1]), color='b', align='center', width=1, facecolor="None", edgecolor='k', linewidth=0.5)
plt.bar((bins_countdurncs[:-1] + bins_countdurncs[1:])/2, countdurncs/len(csevt[csevt['censevt']==0]), color='orange', align='center', width=1, facecolor="None",  edgecolor='k', linewidth=0.5)
plt.legend()
plt.title('Duration', loc='left')
plt.ylabel('Probability')
plt.xlabel('Day')

plt.subplot(3, 1, 2)
countmagcs, bins_countmagcs = np.histogram(csevt[csevt['censevt']==1]['wsmagcs'], bins=16, range=(11,15))
countmagncs, bins_countmagncs = np.histogram(csevt[csevt['censevt']==0]['wsmagcs'], bins=16, range=(11,15))
plt.bar((bins_countmagcs[:-1] + bins_countmagcs[1:])/2, countmagcs/len(csevt[csevt['censevt']==1]), color='b', align='center', width=0.25, alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-CENS')
plt.bar((bins_countmagncs[:-1] + bins_countmagncs[1:])/2, countmagncs/len(csevt[csevt['censevt']==0]), color='orange', align='center', width=0.25, alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-noCENS')
plt.bar((bins_countmagcs[:-1] + bins_countmagcs[1:])/2, countmagcs/len(csevt[csevt['censevt']==1]), align='center', width=0.25, facecolor="None", edgecolor='k', linewidth=0.5)
plt.bar((bins_countmagncs[:-1] + bins_countmagncs[1:])/2, countmagncs/len(csevt[csevt['censevt']==0]), align='center', width=0.25, facecolor="None", edgecolor='k', linewidth=0.5)
plt.legend()
plt.title('Magnitude', loc='left')
plt.ylabel('Probability')
plt.xlabel('m/s')

plt.subplot(3, 1, 3)
censcsm=csevt[csevt['censevt']==1].groupby(csevt[csevt['censevt']==1].index.month).count()['wsmagcs'].values
censncsm=csevt[csevt['censevt']==0].groupby(csevt[csevt['censevt']==0].index.month).count()['wsmagcs'].values
barscs=plt.bar([0.8,1.8,2.8],[censcsm[-1],censcsm[0],censcsm[1]], width=0.4, color='b', align='center', alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-CENS')
barsncs=plt.bar([1.2,2.2,3.2],[censncsm[-1],censncsm[0],censncsm[1]], width=0.4, color='orange', align='center', alpha=0.5, edgecolor='k', linewidth=0.5, label='CS-noCENS')
plt.bar([0.8,1.8,2.8],[censcsm[-1],censcsm[0],censcsm[1]], width=0.4, facecolor="None", align='center', edgecolor='k', linewidth=0.5)
plt.bar([1.2,2.2,3.2],[censncsm[-1],censncsm[0],censncsm[1]], width=0.4, facecolor="None", align='center', edgecolor='k', linewidth=0.5)
plt.xticks([1,2,3], ['Desember', 'Januari', 'Februari'])
pctcenscs=np.int32(censcsm/(censcsm+censncsm)*100)
pctcensncs=np.int32(censncsm/(censcsm+censncsm)*100)
for a,b,c in zip([0.8,1.8,2.8],[censcsm[-1],censcsm[0],censcsm[1]], [pctcenscs[-1],pctcenscs[0],pctcenscs[1]]):
    plt.text(a,b-5,str(c)+" %", ha='center', c='b')
for a,b,c in zip([1.2,2.2,3.2],[censncsm[-1],censncsm[0],censncsm[1]], [pctcensncs[-1],pctcensncs[0],pctcensncs[1]]):
    plt.text(a,b-5,str(c)+" %", ha='center', c='r')
plt.legend()
plt.title('Jumlah Kejadian tiap Bulan', loc='left')
plt.ylabel('Jumlah kejadian')
plt.bar_label(barscs)
plt.bar_label(barsncs)
plt.ylim([0,80])

fig.subplots_adjust(wspace=0.25, hspace=0.35)
#plt.savefig('figure_output\Fig 2.pdf', bbox_inches='tight')
plt.show()
```
